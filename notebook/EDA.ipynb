This is an analysis of the vehicles in the united states

Importing the data frame of cars in the united states

```python
import pandas as pd
```

```python
df = pd.read_csv('vehicles_us.csv')
```

```python
df.head()
```

Filling in missing values for the model year with the median for each grouping by each model

```python
model_year_medians = df.groupby('model')['model_year'].median()

for model, median in model_year_medians.items():
    df.loc[(df['model'] == model) & (df['model_year'].isnull()), 'model_year'] = median
overall_median = df['model_year'].median()
df['model_year'].fillna(overall_median, inplace=True)
```

```python
df.head()
```

Filling in the missing values for the odometer grouping by the cylinders using the median

```python
odometer_medians = df.groupby('cylinders')['odometer'].median()
for cylinders, median in odometer_medians.items():
    df.loc[(df['cylinders'] == cylinders) & (df['odometer'].isnull()), 'odometer'] = median
```

```python
df.head()
```

Filling in paint color missing values with the word missing

```python
df['paint_color'].fillna('missing', inplace=True)
```
